# Notebooks

A notebook is a series of *cells* that can be either:
- *Markdown* meaning they contain text (in markdown language), or
- *code* meaning they contain code that can be run from within the notebook.

You can select a cell by clicking it. To check whether the cell is markdown or code, check the dropdown menu above:



## Editing cells

To edit a cell, click on it and press <kbd>Enter</kbd>. The border around the cell will turn green, and you will be able to edit it's content.

You can press <kbd>esc</kbd> to exit edit mode.

Alternatively, you can press <kbd>^Enter</kbd> when in edit mode to "run" the cell.
- If the cell you run is a markdown cell, this will render the markdown as nicer text, you can use HTML features in these cells if you like.
- If the cell you run is a code cell, pressing <kbd>^Enter</kbd> will execute the code in that cell and print any output produced.

## Kernel

When you open up a notebook, a kernel will automatically start. This runs in the background, and enables you to execute your code.

It is wise, when writing code in a notebook, to restart the kernel every so often. 



This is because variables will be "remembered" for as long as your kernel is running.

So if you have cells like:

```python
x = 5
```

```python
print(x)
```

```python
x = 3
```

And you run them in a different order to that above, x will contain a different value to what it would if they were run in order. 

This can cause confusion in assignments - your code should work as expected when you run all the cells *in order*.

<em><strong>Note:</strong> The kernel will timeout after a period of no use. It is important that you run your code regularly from top to bottom as you work on it, to ensure that your saved work produces results as expected.
