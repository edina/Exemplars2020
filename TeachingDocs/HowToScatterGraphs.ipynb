# Plotting scatter graphs

This notebook will walk through a few different methods for plotting data.

Usually, libraries used are imported at the top of the notebook in which they are used. In this notebook, each library will be imported at the time of use, so you will know exactly what to import if you copy anything to your own work.

### Data

To demonstrate plotting, we need some data to plot.

Scikit learn have a [range of datasets](https://scikit-learn.org/stable/datasets/index.html) ready to use for practise.

This notebook uses the [iris](https://scikit-learn.org/stable/auto_examples/datasets/plot_iris_dataset.html) dataset from scikit-learn.

```python
### import scikit-learn's pre-installed datasets
import sklearn.datasets as ds
```

```python
# load iris dataset
data_full = ds.load_iris()
# take first two features (for a 2D plot)
data = data_full.data[:,:2]
```

<hr>

# The "basic" way

Libraries used in this example are part of the "base scipy stack" or ["core"](https://www.scipy.org/about.html) which is a collection of very common libraries used in data science with python.

Benefits:
- Lots of documentation to help if things go wrong
- Consistent - once you've learned scatter, learning other kinds of plots won't be difficult

Drawbacks:
- Can be more clunky to write
- Less visually appealing graphs than some libraries
- Not interactive

```python
### import matplotlib's plotting functions
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# create figure, axes object
fig1, ax1 = plt.subplots()

# plot data
ax1.scatter(data[:,0], data[:,1])

# add axes labels (using information from the dataset)
ax1.set_xlabel(data_full.feature_names[0])
ax1.set_ylabel(data_full.feature_names[1])

# add title
ax1.set_title('Sepal length and sepal width from Iris data');
```

## Customisation

It is possible to change the appearance of this plot. Look at the following example, with this in mind:
- s = size of markers
- c = colour of markers
- marker = style of markers
- alpha = transparency of markers (0 being translucent, 1 being opaque)

```python
# create figure, axes object as before
fig2, ax2 = plt.subplots()

# plot data
ax2.scatter(data[:,0], data[:,1], s=100, c='red', marker='*', alpha=0.3)

# add axes labels (using information from the dataset)
ax2.set_xlabel(data_full.feature_names[0])
ax2.set_ylabel(data_full.feature_names[1])

# add title
ax2.set_title('Sepal length and sepal width from Iris data');
```

## Deviation - MATLAB style

It is possible to make a quick plot in a style similar to MATLAB.

```python
plt.plot(data[:,0], data[:,1], 'o');
```

# Interactive plotting with Bokeh

[Bokeh](https://bokeh.org) is a library that allows easy creation of interactive plots, without adding much difficulty to creation of the plots.

Benefits:
- Interactive plots (zoom in/drag)

Drawbacks:
- Different vocabulary around graphs ("glyphs")

Once you have run the two cells below, try clicking and dragging the graph to move it. You can also zoom, and interact in other ways by making a selection on the menu at the right of the graph.

```python
### import required bokeh functions
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import bokeh
```

```python
# plot to notebook
output_notebook()        

# create plot figure
p = figure(plot_width = 400, plot_height = 225)

# plot data on figure
p.circle(data[:,0], data[:,1], size=5)

# display plot in notebook
show(p)
```

# labelling points with plotly express

For further documentation on this library, see the [plotly](https://plotly.com/python/getting-started/) website.

Benefits:
- Can hover, drag, zoom in/out - more interactive than the static plot of matplotlib
- Points are labelled, hover over a point to view coordinates.
- Uses vocabulary similar to the base scipy stack.

Drawbacks:
- Can be more overwhelming to new users.
- Styling these plots can become complicated quickly

```python
### import required plotly functions
import plotly.express as px
#import plotly
```

```python
### plot
fig = px.scatter(x=data[:,0], y=data[:,1])
fig.show()
```

# Seaborn

Benefits:
 - you can't see it in this example but it actually looks nicer than with matplotlib (colours, markers, so on)
 
Drawbacks:
- it still kinda uses matplotlib for customizing

```python
### import seaborn
import seaborn as sns
### import matplotlib's plotting functions and set inline
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# set seaborn "style" for graph
sns.set(style="whitegrid")

# plot data with seaborn's plotting function
sns.scatterplot(x= data[:,0], y=data[:,1])

# label x and y axis
plt.xlabel(data_full.feature_names[0])
plt.ylabel(data_full.feature_names[1])

# add title
plt.title('Sepal length and sepal width from Iris data');
```
