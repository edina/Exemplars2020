```python
###TRY BY COPYING AND CHANGING CODE FROM SOMEWHERE
from bokeh.layouts import column
from bokeh.models import CustomJS, ColumnDataSource, Slider

x = np.linspace(0, 8*np.pi, 800)
y = np.sin(x)
source = ColumnDataSource(data=dict(x=x, y=y))

plot = figure(plot_width=400, plot_height=400)
plot.line('x', 'y', source=source, line_width=3, line_alpha=0.6)


update_curve = CustomJS(args=dict(source=source), code="""
    var data = source.data;
    var f = slider.value;
    var x = data['x']
    var y = data['y']
    
    for (var i = 0; i < x.length; i++) {
        y[i] = np.sin(x[i]) + f}
    
    // necessary because we mutated source.data in-place
    source.change.emit();
""")
slider = Slider(title="Offset", value=0.0, start=-5.0, end=5.0, step=0.1)

slider.js_on_change('value', update_curve)

show(column(slider, plot))
```

```python
###TRY WITH MORE WIDGETS IN JS- DOESN'T WORK
# Set up data
x = np.linspace(0, 8*np.pi, 800)
y = np.sin(x)
source = ColumnDataSource(data=dict(x=x, y=y))

# Set up plot
plot = figure(plot_height=400, plot_width=600, title="Sine wave",
              tools="crosshair,pan,reset,save,wheel_zoom",
              x_range=[0, 8*np.pi], y_range=[-2.5, 2.5])

plot.line('x', 'y', source=source, line_width=3, line_alpha=0.6)

# Set up widgets
text = TextInput(title="Title", value='Sine wave')
offset = Slider(title="Offset", value=0.0, start=-5.0, end=5.0, step=0.1)
amplitude = Slider(title="Amplitude", value=1.0, start=-5.0, end=5.0, step=0.1)
phase = Slider(title="Phase", value=0.0, start=0.0, end=2*np.pi)
freq = Slider(title="Frequency", value=1.0, start=0.1, end=5.1, step=0.1)
trig = Dropdown(label="Trigonometric equation",
                menu=[('sin', 'sin'), ('cos', 'cos')], value='sin')

update_title = CustomJS(args=dict(source=source, slider=text), code="""
    plot.title.text = text.value""")

text.js_on_change('value', update_title)

update_curve = CustomJS(args=dict(source=source, widgets=[offset, amplitude, phase, freq, trig]), code="""
    /// Get the current slider values
    var data = source.data
    var a = amplitude.value
    var b = offset.value
    var w = phase.value
    var k = freq.value
    var t = trig.value

    /// Generate the new curve
    var x = np.linspace(0, 8*np.pi, 400)
    if var t == 'sin':
        var y = a*np.sin(k*x + w) + b
        var data = dict(x=x, y=y)
    else:
        var y = a*np.cos(k*x + w) + b
        var data = dict(x=x, y=y)
    source.change.emit()""")

for w in [offset, amplitude, phase, freq, trig]:
    w.js_on_change('value', update_curve)

# Set up layouts and add to document
inputs = column(trig, text, offset, amplitude, phase, freq)
layout = row(inputs, plot, width=800)
show(layout)
```

## Simple line plot

```python
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook
import numpy as np
x = np.linspace(-10, 10, 21)
y = x**2 + 2

output_file("multiple.html")

p = figure(plot_width=400, plot_height=400)

# add both a line and circles on the same plot
p.line(x, y, line_width=2)
p.circle(x, y, fill_color="white", size=8)
output_notebook()
show(p)
```
